## Business Problem Statement

Imagine you are business man that is looking to start a new business in a given city. One of the first things you need to do is to identify the best location to establish your business. In order to do that, you need to study the current market and demand for the particular kind of business you are interested in establishing on the city. This task may represent a real challenge, specially if you are new to the city and you are not familiar with the different neighborhoods and their distribution around in the city. 


## Proposed Solution 

We proposed to create a tool to assist the users to analyze a given city's blueprint in regards of a particular kind of business. This is achived by generating a map of the city and pin-pointing the most popular venues of the kind of business in scope. The map, in turn, gives the users a city-wide overview of the venues distribution, which may be used to identify areas with no venues or areas with too many venues. This information results very usefull to decide on the best location to establish a new business venue.

### Visualize the map

```python
# create map
map_venues = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to the map
markers_color = colors.rgb2hex(colors.BASE_COLORS['r'])
for lat_i, lon_i, name_i, cat_i in zip(venues_df['Latitude'], venues_df['Longitude'], venues_df['Venue'], venues_df['Category']):
    label = folium.Popup(str(name_i) + ' - Category: ' + str(cat_i), parse_html=True)
    folium.CircleMarker([lat_i, lon_i], radius=5, popup=label, color=markers_color, fill=True,
                        fill_color=markers_color, fill_opacity=1).add_to(map_venues)

map_venues
```
